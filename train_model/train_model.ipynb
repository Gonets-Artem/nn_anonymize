```python
model = YOLO("yolov8n-seg.pt")
```

```python
model.train(
    data='/kaggle/input/video-conference-argon/dataset_video_conference_3/dataset_video_conference_3/data.yaml',  # Путь к файлу конфигурации с данными
    epochs=80,                          # Количество эпох
    batch=32,                           # Размер батча
    lr0=0.01,                           # Начальная скорость обучения
    lrf=0.2,                            # Коэффициент для уменьшения lr
    warmup_epochs=5,           # Количество итераций для warmup
    iou=0.5,          # Порог IoU
    conf=0.25,     # Порог уверенности
    device='0',                         # Указываем GPU (если доступен)
    project='/kaggle/working/save/models',     # Папка для сохранения модели
    name='yolov8-videoconference',            # Имя проекта
    exist_ok=True,                      # Если проект существует, продолжим обучение
)
```

```python
model.save('my_model_6_ds3.pt')
```

```python
import shutil

# Укажите путь к вашей папке
folder_path = '/kaggle/working/save'
output_path = '/kaggle/working/my_archive_6_ds3.zip'

# Создание архива
shutil.make_archive(output_path.replace('.zip', ''), 'zip', folder_path)
```

```python
save_path = '/kaggle/working/predictions'
os.makedirs(save_path, exist_ok=True)
```

```python
import shutil


test_images_path = '/kaggle/input/video-conference-argon/dataset_video_conference_3/dataset_video_conference_3/test/images'
s = os.listdir(test_images_path)

for img in s:
    results = model.predict(os.path.join(test_images_path, img))[0]
    save_file = os.path.join(save_path, img)
    results.save(filename=save_file)

# Укажите путь к вашей папке
folder_path = '/kaggle/working/predictions'
output_path = '/kaggle/working/my_archive_predictions_6_ds3.zip'

# Создание архива
shutil.make_archive(output_path.replace('.zip', ''), 'zip', folder_path)
```
