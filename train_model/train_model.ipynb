```python
model = YOLO("yolov8n-seg.pt")
```

```python
model.train(
    data='/kaggle/input/video-conference-argon/dataset_video_conference_3/dataset_video_conference_3/data.yaml',  # ĞŸÑƒÑ‚ÑŒ Ğº Ñ„Ğ°Ğ¹Ğ»Ñƒ ĞºĞ¾Ğ½Ñ„Ğ¸Ğ³ÑƒÑ€Ğ°Ñ†Ğ¸Ğ¸ Ñ Ğ´Ğ°Ğ½Ğ½Ñ‹Ğ¼Ğ¸
    epochs=80,                          # ĞšĞ¾Ğ»Ğ¸Ñ‡ĞµÑÑ‚Ğ²Ğ¾ ÑĞ¿Ğ¾Ñ…
    batch=32,                           # Ğ Ğ°Ğ·Ğ¼ĞµÑ€ Ğ±Ğ°Ñ‚Ñ‡Ğ°
    lr0=0.01,                           # ĞĞ°Ñ‡Ğ°Ğ»ÑŒĞ½Ğ°Ñ ÑĞºĞ¾Ñ€Ğ¾ÑÑ‚ÑŒ Ğ¾Ğ±ÑƒÑ‡ĞµĞ½Ğ¸Ñ
    lrf=0.2,                            # ĞšĞ¾ÑÑ„Ñ„Ğ¸Ñ†Ğ¸ĞµĞ½Ñ‚ Ğ´Ğ»Ñ ÑƒĞ¼ĞµĞ½ÑŒÑˆĞµĞ½Ğ¸Ñ lr
    warmup_epochs=5,           # ĞšĞ¾Ğ»Ğ¸Ñ‡ĞµÑÑ‚Ğ²Ğ¾ Ğ¸Ñ‚ĞµÑ€Ğ°Ñ†Ğ¸Ğ¹ Ğ´Ğ»Ñ warmup
    iou=0.5,          # ĞŸĞ¾Ñ€Ğ¾Ğ³ IoU
    conf=0.25,     # ĞŸĞ¾Ñ€Ğ¾Ğ³ ÑƒĞ²ĞµÑ€ĞµĞ½Ğ½Ğ¾ÑÑ‚Ğ¸
    device='0',                         # Ğ£ĞºĞ°Ğ·Ñ‹Ğ²Ğ°ĞµĞ¼ GPU (ĞµÑĞ»Ğ¸ Ğ´Ğ¾ÑÑ‚ÑƒĞ¿ĞµĞ½)
    project='/kaggle/working/save/models',     # ĞŸĞ°Ğ¿ĞºĞ° Ğ´Ğ»Ñ ÑĞ¾Ñ…Ñ€Ğ°Ğ½ĞµĞ½Ğ¸Ñ Ğ¼Ğ¾Ğ´ĞµĞ»Ğ¸
    name='yolov8-videoconference',            # Ğ˜Ğ¼Ñ Ğ¿Ñ€Ğ¾ĞµĞºÑ‚Ğ°
    exist_ok=True,                      # Ğ•ÑĞ»Ğ¸ Ğ¿Ñ€Ğ¾ĞµĞºÑ‚ ÑÑƒÑ‰ĞµÑÑ‚Ğ²ÑƒĞµÑ‚, Ğ¿Ñ€Ğ¾Ğ´Ğ¾Ğ»Ğ¶Ğ¸Ğ¼ Ğ¾Ğ±ÑƒÑ‡ĞµĞ½Ğ¸Ğµ
)
```

```python
model.save('my_model_6_ds3.pt')
```

```python
save_path = '/kaggle/working/predictions'
os.makedirs(save_path, exist_ok=True)
```

```python
import shutil


test_images_path = '/kaggle/input/video-conference-argon/dataset_video_conference_3/dataset_video_conference_3/test/images'
s = os.listdir(test_images_path)

for img in s:
    results = model.predict(os.path.join(test_images_path, img))[0]
    save_file = os.path.join(save_path, img)
    results.save(filename=save_file)

# Ğ£ĞºĞ°Ğ¶Ğ¸Ñ‚Ğµ Ğ¿ÑƒÑ‚ÑŒ Ğº Ğ²Ğ°ÑˆĞµĞ¹ Ğ¿Ğ°Ğ¿ĞºĞµ
folder_path = '/kaggle/working/predictions'
output_path = '/kaggle/working/my_archive_predictions_6_ds3.zip'

# Ğ¡Ğ¾Ğ·Ğ´Ğ°Ğ½Ğ¸Ğµ Ğ°Ñ€Ñ…Ğ¸Ğ²Ğ°
shutil.make_archive(output_path.replace('.zip', ''), 'zip', folder_path)
```
